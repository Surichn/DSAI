Visualization using Seaborn

```python
# importing required packages

import pandas as pd
from matplotlib import pyplot as pp
import seaborn as sb
```

```python
url = "https://raw.githubusercontent.com/Surichn/DSAI/main/DA/Taxi_Trips.csv"
df = pd.read_csv(url)
```

```python
print(f"Inference Data : {df.info()}")
```

```python
print (df.head(10))
```

```python
sb.violinplot(x="payment_type", y="total_amount", data=df)
```

```python
sb.boxplot(x="payment_type", y ="total_amount", data=df)
```

```python
g = sb.PairGrid(df,hue="payment_type", palette = "Oranges")
g.map_diag(pp.hist)
g.map_offdiag(pp.scatter)
pp.show()
```
